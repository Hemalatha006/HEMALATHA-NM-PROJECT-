```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
df = pd.read_csv('/kaggle/input/canteen-shop-transaction-data/canteen_shop_data.csv')

# Display the first few rows of the DataFrame
print(df.head())

# Basic statistics of numerical columns
print(df.describe())

# Visualize total sales per day
daily_sales = df.groupby('Date')['Total'].sum()
plt.figure(figsize=(10, 6))
daily_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales per Day')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.show()

# Visualize total sales by payment method
payment_sales = df.groupby('Payment Method')['Total'].sum()
plt.figure(figsize=(8, 5))
payment_sales.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=140)
plt.title('Total Sales by Payment Method')
plt.ylabel('')
plt.show()

# Visualize customer satisfaction distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Customer Satisfaction', palette='Set2')
plt.title('Customer Satisfaction Distribution')
plt.xlabel('Satisfaction Level')
plt.ylabel('Count')
plt.show()
```
